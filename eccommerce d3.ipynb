```python
import pandas as pd
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.sentiment import SentimentIntensityAnalyzer
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/drive/MyDrive/cleaned_data.csv')
```

```python
import nltk
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('vader_lexicon')
```

```python
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

df['tokenized_text'] = df['reviews.text'].apply(word_tokenize)
df['filtered_text'] = df['tokenized_text'].apply(lambda x: [lemmatizer.lemmatize(word) for word in x if word not in stop_words])
```

```python
sia = SentimentIntensityAnalyzer()

df['sentiment_scores'] = df['filtered_text'].apply(lambda x: sia.polarity_scores(' '.join(x)))
```

```python
df['sentiment_scores']
```

```python
#df['sentiment'] = df['sentiment_scores'].apply(lambda x: 0 if x['compound'] >= 0 else 1)
#df_filtered['sentiment'].replace({'negative': 0, 'positive': 1}, inplace=True)
```

```python
df['sentiment'] = df['sentiment_scores'].apply(lambda x: 'negative' if x['compound'] < 0 else 'positive')
# df_filtered['sentiment'].replace({'negative': 0, 'positive': 1}, inplace=True)
```

```python
df['sentiment'] = df['sentiment'].reset_index(drop=True)
```

```python
spark = SparkSession.builder.getOrCreate()
```

```python
# Load the dataset into a Spark DataFrame
df = spark.read.csv('/content/drive/MyDrive/cleaned_data.csv', header=True, inferSchema=True)
```

```python

# Define UDFs for tokenization and lemmatization
tokenizer_udf = udf(lambda x: word_tokenize(x), ArrayType(StringType()))
lemmatizer_udf = WordNetLemmatizer()
```

```python
# Tokenize and lemmatize the text data
df = df.withColumn('tokenized_text', tokenizer_udf('reviews.text'))
df = df.withColumn('filtered_text', udf(lambda x: [lemmatizer_udf.lemmatize(word) for word in x if word not in stop_words], ArrayType(StringType()))('tokenized_text'))
```

```python
# Remove stop words
stop_words_remover = StopWordsRemover(inputCol='filtered_text', outputCol='filtered_text_no_stop_words')
df = stop_words_remover.transform(df)
```
